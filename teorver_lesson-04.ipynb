#### Задание №1
Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].
Найдите ее среднее значение и дисперсию

1-й способ: Интегрированием в scipy формул $M(X) = \displaystyle\int_{-\infty}^\infty x \cdot f(x) \: dx$ и $D(X) = \displaystyle\int_{-\infty}^\infty (x - M(X))^2 \cdot f(x) \: dx$

```python
from scipy import integrate
def f(x):
    return 1/(b-a)*x
a=200
b=800

M, err = integrate.quad(f, a, b)
def f2(x):
    return (x-M)**2/(b-a)

D, err = integrate.quad(f2, a, b)
print (f"Мат.ожидание M={round(M)}, Дисперсия D={round(D)}")
```

2-й способ: Проинтегрировав в общем виде (от a до b) формулы определяющие мат.ожидание и дисперсию для равномерного распределения ($f(x)=1/(b-a)$) получим:
$M(X) = (a + b)/2$ и  $D(X) = (b − a)^2/12$

Мат.ожидание M=500, Дисперсия D=30000

#### Задание №4
Рост взрослого населения города N имеет нормальное распределение.
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.
Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
а). больше 182 см
б). больше 190 см
в). от 166 см до 190 см
г). от 166 см до 182 см
д). от 158 см до 190 см
е). не выше 150 см или не ниже 190 см
ё). не выше 150 см или не ниже 198 см
ж). ниже 166 см.

Обозначим вероятность как $$P(a < X < b) = F(b) - F(a)$$
где $F(x)$ - значения нормального распределения, $a$ и $b$ границы интервала вероятность попадания в который мы ищем.


```python
from scipy import stats
norm = stats.norm(loc=174, scale=8)

print (f"a). x>182 Вероятность P={1-norm.cdf(182)}")
print (f"б). x>190 Вероятность P={1-norm.cdf(190)}")
print (f"в). 166<x<190 Вероятность P={norm.cdf(190)-norm.cdf(166)}")
print (f"г). 166<x<182 Вероятность P={norm.cdf(182)-norm.cdf(166)}")
print (f"д). 158<x<190 Вероятность P={norm.cdf(190)-norm.cdf(158)}")
print (f"е). x<150 или x>190 Вероятность P={1-(norm.cdf(190)-norm.cdf(150))}")
print (f"ё). x<150 или x>198 Вероятность P={(1-norm.cdf(198))+norm.cdf(150)}")
print (f"ж). x<166 Вероятность P={norm.cdf(166)-0}")
```
